# Step 3:Data Preparation

```python
#Dropping columns that are irrelevant for the model

bike_sharing.drop(['instant','dteday','casual','registered'],axis = 1,inplace = True)
bike_sharing.head()
```

```python
#Converting numerical values for columns 'weathersit' and 'season' into categorical string values

bike_sharing.season=bike_sharing['season'].replace([1,2,3,4],['Spring','Summer','Fall','Winter'])

bike_sharing.weathersit=bike_sharing['weathersit'].map({1:'Clear',2:'Mist & Cloudy',3:'Light Snow & Rain',4:'Heavy Snow & Rain'})
```

```python
#Cross checking values in season

bike_sharing['season'].value_counts()
```

```python
#Cross checking values in weathersit

bike_sharing['weathersit'].value_counts()
```

```python
#Converting 'mnth' numerical value column into categorical values

import calendar
bike_sharing['mnth']=bike_sharing['mnth'].apply(lambda x: calendar.month_abbr[x])
```

```python
#Validating month column

bike_sharing['mnth'].value_counts()
```

```python
#Converting weekday values into categorical type

bike_sharing.weekday = bike_sharing.weekday.apply(lambda x:calendar.day_name[x])
```

```python
#Validating the weekdays

bike_sharing.weekday.value_counts()
```

```python
# Now let's see the head of our dataframe.

bike_sharing.head()
```

```python
#Checking info of dataset

bike_sharing.info()
```

Thus,we have converted the numerical values of columns of 'season','weathersit','weekday' and 'mnth' into categorical string values

# Dummy Variables

```python
#Creating dummy variables for season , mnth ,weathersit ,weekday

dummy=bike_sharing[['season','mnth','weekday','weathersit']]
```

```python
dummy=pd.get_dummies(dummy)
```

```python
# Check what the dataset 'dummy' looks like

dummy.head()
```

```python
# Let's drop the first column from status df using 'drop_first = True'

dummy=pd.get_dummies(dummy,drop_first=True)
```

```python
#Adding dummy variables to original dataset

bike_sharing=pd.concat([dummy,bike_sharing],axis = 1)
```

```python
# Now let's see the head of our dataframe.

bike_sharing.head()
```

```python
#Dropping season,mnth,weathersit,weekday columns for which the dummy variables were created

bike_sharing=bike_sharing.drop(['season','mnth','weathersit','weekday'],axis=1)
```

```python
bike_sharing.shape
```

# Step 4:Splitting the Data into Training and Testing Sets

As you know, the first basic step for regression is performing a train-test split.

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)

train,test=train_test_split(bike_sharing,train_size=0.7,test_size=0.3,random_state=100)
```

# Step 5:Rescaling the Features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'categoriacl variables' and 'dummy' variables

num_vars = ['cnt', 'hum', 'windspeed', 'temp', 'atemp']

train[num_vars] = scaler.fit_transform(train[num_vars])
```

```python
train.head()
```

```python
train.describe()
```

Thus,we have min-max values of numeric varaibles ranging between 0 and 1 after scaling

```python
#Let's check the correlation coefficients to see which variables are highly correlated
plt.figure(figsize=(22,10))
sns.heatmap(train.corr(),annot=True,cmap='YlGnBu')
plt.show()
```

Note:I observed that cnt is highly varaibles atemp based on above correlation

```python
plt.figure(figsize=[6,6])
plt.scatter(train.atemp, train.cnt)
plt.show()
```

We'll first build a model using all the columns

# Dividing into X and Y sets for the model building

```python
y_train=train.pop('cnt')
X_train=train
```

# Step 6:Building a linear model

```python
#We will be using Linear Regression function from SciKit learn for its compatibility with RFE
```

# Model 1:

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#Running RFE with the output number of the variable equal to 13

lm=LinearRegression()
lm.fit(X_train,y_train)

rfe=RFE(lm,13)
rfe=rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col=X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

# Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables

X_train_rfe=X_train[col]
```

```python
# Adding a constant variable

import statsmodels.api as sm

X_train_rfe=sm.add_constant(X_train_rfe)
```

```python
#Running the linear model

lm=sm.OLS(y_train,X_train_rfe).fit()
```

```python
# Check the parameters obtained

lm.params
```

```python
#Let's see the summary of our linear model

print(lm.summary())
```

Note:atemp is insignificant in comparison with other variables and can be dropped

```python
X_train_new = X_train_rfe.drop(["atemp"], axis = 1)
```

# Step 7:Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_cnt=lm.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms

fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 15)
plt.xlabel('Errors', fontsize = 15)
```

# Step 8:Making Predictions

Now that we have fitted the model and checked the normality of error terms, it's time to go ahead and make predictions using the final model

```python
# Making predictions
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
